```python
import pandas as pd

# para pintar
import pylab as plt
import seaborn as sns

# para que salga el grafico
%matplotlib inline

pathData = "/Users/carlosbasco/Desktop/Projects/Data-Cleaning-Pandas-Project/data/sharks_clean.csv"
df = pd.read_csv(pathData, encoding= "UTF8")
```

```python
##Sucesos por año -------------------------------------------------------------------- #

plt.figure(figsize=(15,10))

df.Year.hist(bins=300)

plt.xlabel('Year')
plt.ylabel('Count');
```

```python
###En los últimos 50 años -------------------------------------------------------------------- #

subSet = df[df["Year"]>1970]
plt.figure(figsize=(15,10))

subSet.Year.hist(bins=50)

plt.xlabel('Year')
plt.ylabel('Count');
```

```python
##Sucesos por meses -------------------------------------------------------------------- #

plt.figure(figsize=(15,10))

#df.Date.hist(bins=40)

df.Date.value_counts().plot.bar(rot = 360);

plt.ylabel('Count');

```

```python
##Sucesos por pais -------------------------------------------------------------------- #

plt.figure(figsize=(15,10))

df.Country.value_counts().plot.bar();
plt.ylabel('Count');
```

```python
###Sucesos por pais en los últimos 20 años ----------------------------------------------------- #

plt.figure(figsize=(15,10))

df.Country[df.Year>2002].value_counts().plot.bar();

plt.ylabel('Count');
```

```python
##Sucesos por actividad ----------------------------------------------------- #
plt.figure(figsize=(15,8))

df.Activity.value_counts().plot.bar(rot=45);
plt.ylabel('Count');
```

```python
##Sucesos por sexo ----------------------------------------------------- #
plt.figure(figsize=(15,7))

df.Sex.value_counts().plot.bar(rot = 360);

plt.xlabel('Sex')
plt.ylabel('Count');
```

```python
##Especie de tiburon por caso ----------------------------------------------------- #
plt.figure(figsize=(15,8))

df.Species.value_counts().plot.bar();
plt.ylabel('Count');
```

```python
###Casos con especies identificadas -------------------------------------------- #
plt.figure(figsize=(15,8))

dfAux = df[df.Species != "Undefined shark"]
dfAux.Species.value_counts().plot.bar(rot=360);
plt.ylabel('Count');
```
