```python
df["Case Number"].value_counts() #Vamos a cambiar los 0s y el "xx" dado que no tenemos informacion realmente
```

```python
df["Case Number"] = df["Case Number"].replace(r"[A-Z]{2}.[0-9]{4}|.[a-zA-Z]|[0-9]{4}.[0-9]{4}", value = "", regex = True)
df["Case Number"] = df["Case Number"].replace("", value = None)
df["Case Number"] = df["Case Number"].replace('0', value = None)
df["Case Number"] = df["Case Number"].replace('0500.00.00', value = None) 
df["Case Number"] = df["Case Number"].replace('0077.00.00', value = None)
df["Case Number"] = df["Case Number"].replace('0005.00.00', value = None)

#df["Case Number"] = df["Case Number"].replace('\.', value = "-", regex = True)
```

```python
df["Date"].value_counts() #Vamos a limpiar lo más que podamos la data
```

```python
df["Date"] = df["Date"].replace(r"Reported *|M\n|Early|Late|Ca\.* *|Fishing|\\n|Beforer|Before|No date|or|Reportd|Reprted|Woirld War II| but took place around 1868|[Cc]irca|\"Anniversary Day\"|Between|A few years before|World War II|\"During the war\"|\"Before the war\"| late|Letter dated|-* *\d{1,4}-* *", value = "", regex = True)
df["Date"] = df["Date"].replace(r"\!|\"|\#|\$|\%|\&|\'|\(|\)|\*|\+|\,|\\|\-|\.|\/|\:|\;|\<|\=|\>|\?|\@|\[|\]|\^|\_|\`|\{|\||\}", value = "", regex = True)
df["Date"] = df["Date"].replace(r"[a-zA-Z]* ", value = "", regex = True)
df["Date"] = df["Date"].replace(r"[A-Z]{2}|R|a*sf*", value = "", regex = True)
df["Date"] = df["Date"].replace(['[Ww]inter', '[Ss]*ummer','[Ff]all'], ["Dic","Jun","Sep"], regex = True) #value=)
```
